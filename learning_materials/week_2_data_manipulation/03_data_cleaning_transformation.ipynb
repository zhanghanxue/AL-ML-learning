```python
import pandas as pd

titanic_df = pd.read_csv("../titanic/train.csv")
```

```python
titanic_df.isnull().sum()
```

```python
unique_cabins = titanic_df["Cabin"].dropna().unique()
unique_cabins_count = len(unique_cabins)
print(unique_cabins)
print(unique_cabins_count)
```

```python
multi_cabin_entries = [cabin for cabin in unique_cabins if " " in cabin]
print(multi_cabin_entries)
print(f"Number of cabins with multiple values: {len(multi_cabin_entries)}")
```

```python
titanic_df_cleaning = titanic_df.copy()
titanic_df_cleaning = titanic_df_cleaning.dropna(subset=["Cabin", "Embarked"])

titanic_df_cleaning.isnull().sum()
```

```python
titanic_df_cleaning.fillna({"Age": titanic_df["Age"].median()}, inplace=True)
```

```python
titanic_df_cleaning
```

```python
titanic_df_cleaning["FamilySize"] = (
    titanic_df_cleaning["SibSp"] + titanic_df_cleaning["Parch"] + 1
)
titanic_df_cleaning["IsAlone"] = titanic_df_cleaning["FamilySize"] == 1
titanic_df_cleaning.head()
```

```python
titanic_df_cleaning["AgeGroup"] = pd.cut(
    titanic_df_cleaning["Age"],
    bins=[0, 12, 20, 40, 60, 80],
    labels=["Child", "Teenager", "Adult", "MiddleAged", "Senior"],
)

titanic_df_cleaning
```

```python
titanic_df_cleaning["AgeGroup"].value_counts()
```

```python
ticket_price = titanic_df[["PassengerId", "Ticket", "Fare"]].copy()

ticket_price
```

```python
titanic_df.drop(columns=[["Fare", "Ticket"]], inplace=True)

titanic_df.head()
```

```python
ticket_price.head()
```

```python
merged_titanic_df = pd.merge(titanic_df, ticket_price, on="PassengerId", how="left")

merged_titanic_df.head()
```

```python
pivot = merged_titanic_df.pivot_table(
    index="Sex", columns="Pclass", values="Survived", aggfunc="mean"
)
print(pivot)
print((pivot * 100).round(1))  # percentages
```

```python
melt_pivot = pivot.melt()
print(melt_pivot)
```

```python
p_false = titanic_df_cleaning.pivot_table(
    index=["AgeGroup", "Sex"],
    columns="Pclass",
    values="Survived",
    aggfunc="mean",
    observed=False,
)
p_true = titanic_df_cleaning.pivot_table(
    index=["AgeGroup", "Sex"],
    columns="Pclass",
    values="Survived",
    aggfunc="mean",
    observed=True,
)
print("shape observed=False:", p_false.shape)
print("shape observed=True :", p_true.shape)

print(p_false)
print("-----")
print(p_true)
```

```python
df = pd.DataFrame(
    {
        "A": pd.Categorical(
            ["x", "x", "y"], categories=["x", "y", "z"]
        ),  # 'z' category never appears
        "B": ["u", "v", "u"],
        "val": [1, 0, 1],
    }
)

print(df)

# observed=False (shows z in the product)
pt_false = df.pivot_table(
    index="A", columns="B", values="val", aggfunc="mean", observed=False
)
# observed=True (omits combinations involving 'z' because it's never observed)
pt_true = df.pivot_table(
    index="A", columns="B", values="val", aggfunc="mean", observed=True
)

print("\nobserved=False:\n", pt_false)
print("\nobserved=True:\n", pt_true)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

titanic = pd.read_csv("../titanic/train.csv")
```

```python
sns.scatterplot(data=titanic, x="Age", y="Fare", hue="Survived")
```

```python
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(data=titanic, x='Age', y='Fare', hue='Survived')

# Add titles and labels
plt.title('Titanic: Fare vs Age by Survival Status', fontsize=14, fontweight='bold')
plt.xlabel('Age (Years)', fontsize=12)
plt.ylabel('Fare (Currency Units)', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# Select only the numeric columns for correlation
numeric_data = titanic[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']]

# Calculate the correlation matrix
correlation_matrix = numeric_data.corr()

# Create the heatmap
sns.heatmap(correlation_matrix, 
            annot=True,        # Annotates the cells with the correlation value
            cmap='coolwarm',   # Uses a red-blue color scheme
            center=0,          # Centers the colormap at zero
            square=True,       # Makes the cells square
            fmt=".2f")         # Formats the annotations to 2 decimal places
```

```python
plt.figure(figsize=(10, 8))
numeric_data = titanic[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].corr()

# Create heatmap with annotations
heatmap = sns.heatmap(numeric_data, annot=True, cmap='coolwarm', center=0, fmt=".2f")

# Add title
plt.title('Titanic Dataset Correlation Heatmap', fontsize=14, fontweight='bold', pad=20)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)
plt.yticks(rotation=0)

plt.tight_layout()
plt.show()
```

```python
p = sns.pairplot(data=titanic, hue="Survived", diag_kind="kde", markers=["o", "s"],
             vars=["Age", "Fare", "Pclass", "SibSp", "Parch"])

for ax in p.axes.flat:
    ax.set_xlabel(ax.get_xlabel(), visible=True)
    ax.set_ylabel(ax.get_ylabel(), visible=True)

plt.suptitle('Pairplot of Titanic Dataset', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Create pairplot and get the grid object
g = sns.pairplot(data=titanic, hue="Survived", diag_kind="kde", 
                 vars=["Age", "Fare", "Pclass", "SibSp", "Parch"],
                 plot_kws={'alpha': 0.7})

# Main title
g.fig.suptitle('Pairplot of Titanic Numeric Variables by Survival', 
               fontsize=16, fontweight='bold', y=1.02)

# Make all axis labels visible
for ax in g.axes.flat:
    if ax.get_xlabel():  # Only if there's an x-label
        ax.set_xlabel(ax.get_xlabel(), visible=True, fontsize=10)
    if ax.get_ylabel():  # Only if there's a y-label
        ax.set_ylabel(ax.get_ylabel(), visible=True, fontsize=10)

# Customize legend
g._legend.set_title('Survived')
for t, l in zip(g._legend.texts, ['No', 'Yes']):
    t.set_text(l)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

df = pd.read_csv('../titanic/train.csv')
```

```python
df.head(10)
```

```python
print(df.groupby('Pclass').Age.mean())
```

```python
df_age_means = df.groupby('Pclass', as_index=False)['Age'].mean()
print(df_age_means)
```

```python
df_grouped_agg = df.groupby('Pclass')['Age'].agg(['mean','median','count'])
print(df_grouped_agg)
```

```python
df_mins = df.groupby('Embarked', as_index=False)['Fare'].min()
df_max = df.groupby('Embarked', as_index=False)['Fare'].max()
df_means = df.groupby('Embarked', as_index=False)['Fare'].mean()

print(df_mins)
print(df_max)
print(df_means)
```

```python
total_zero = (df['Fare'] == 0).sum()
by_embarked = df.loc[df['Fare'] == 0, ('Embarked', 'Pclass','Fare')].value_counts(dropna=False)
print("Total zero fares:", total_zero)
print(by_embarked)
```

```python
# compute mean and median
mean_fare = df['Fare'].mean()
median_fare = df['Fare'].median()
print("Fare mean:", mean_fare)
print("Fare median:", median_fare)

# compare effect of imputing with mean vs median for Age
df_mean_imputed = df.copy()
df_mean_imputed['Age'] = df_mean_imputed['Age'].fillna(df_mean_imputed['Age'].mean())
print(df_mean_imputed['Age'].head(10))

df_median_imputed = df.copy()
df_median_imputed['Age'] = df_median_imputed['Age'].fillna(df_median_imputed['Age'].median())
print(df_median_imputed['Age'].head(10))
```
